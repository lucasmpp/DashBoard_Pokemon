```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import numpy as np
pokemon = pd.read_csv (r'F:\UnB.2\CE2\pokemon.csv')
pokemon.head()
```

```python
def chart_egg_x_tot(df):
    x = df.base_egg_steps
    y = df.base_total
    fig = px.box(df, x=x, y=y, hover_name=pokemon.name, category_orders={"base_egg_steps":["1280","2560","3840","5120","6400",
                                                                                           "7680","8960","10240","20480","30720"]})
    fig.update_xaxes(title_text='Número de passos para chocar o ovo')
    fig.update_yaxes(title_text='Atributos totais')
    fig.update_xaxes(type='category')
    return fig

def chart_egg_x_catch(df):
    a = list(df.capture_rate)
    a[773]=0
    y = list(map(int, a))
    y[773]=np.nan
    x = df.base_egg_steps
    fig = px.box(df, x=x, y=y, hover_name=pokemon.name, category_orders={"base_egg_steps":["1280","2560","3840","5120","6400",
                                                                                           "7680","8960","10240","20480","30720"]})
    fig.update_xaxes(title_text='Número de passos para chocar o ovo')
    fig.update_yaxes(title_text='Facilidade de captura')
    fig.update_xaxes(type='category')
    return fig

def chart_catch_x_tot(df):
    a = list(df.capture_rate)
    a[773]=0
    x = list(map(int, a))
    x[773]=np.nan
    y=df.base_total
    fig = px.scatter(df, x=x, y=y,hover_name=pokemon.name)
    fig.update_xaxes(title_text='Facilidade de captura')
    fig.update_yaxes(title_text='Atributos totais')
    return fig
```

```python
chart_egg_x_tot(pokemon)
```

```python
chart_catch_x_tot(pokemon)
```
